```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data_features_added.csv')
df['date'] = pd.to_datetime(df['date'])
df
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes_flat = axes.flatten()

for i, ax in enumerate(axes_flat):
    col = f'store_id_{i+1}'
    store_data = df[df[col] == 1]
    
    for item in store_data['item_id'].unique():
        item_data = store_data[store_data['item_id'] == item]
        ax.plot(item_data['date'], item_data['item_count'], label=item)
        ax.legend()
```
